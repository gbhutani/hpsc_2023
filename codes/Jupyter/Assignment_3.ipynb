f(x) = e^x = 1+x+(x^2/2!)+(x^3/3!)+………….+(x^n/n!) 

f(x) = e^x ≈ 1+x+(x^2/2!)+(x^3/3!)+………….+(x^100/100!) [for default case]

```python
# first writing factorial function using recusrion

def fac(x):
    if x ==0:
        return 1
    else :
        return x*fac(x-1)   
```

```python
# lets verify the factorial function
fac(5),fac(0),fac(1)
```

```python
#now lets write exponential function
def myexp(x,n=100):
    sum_t = 1
    tol = 1e-12
    for i in range(1,n+1):
        sum_p = sum_t
        sum_t=sum_t+(x**i/fac(i))
        delta_s = sum_t - sum_p
        if delta_s < tol:
            break
#     print(f"After {i+1} iteration, exponetial of {x} is = {sum_t}")
    return sum_t
        
```

```python
myexp(15)
```

```python
# Now plotting numpy exp() and myexp() functions
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# make data
x = np.linspace(0,100,1000)
ynp= np.exp(x)
ymy = np.zeros((1000))
for i in range(1000):
    ymy[i] = myexp(x[i],125)

# plot 
plt.figure(figsize=(12,7))
plt.plot(x,ynp,linewidth=2.0,label="numpy exp() func",marker="o",markerfacecolor ="Red")
plt.plot(x,ymy,linewidth=2.0,label="My exp() func",marker="o",markerfacecolor ="Black")
plt.xlabel("x")
plt.ylabel("exp(x)")
plt.legend()
plt.show()
```
