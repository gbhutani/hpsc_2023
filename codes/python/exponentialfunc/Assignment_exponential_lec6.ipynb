```python
#creating function for exponential series
def your_exp(x,n=100):
    a=x
    b=1
    t_old=1
    for i in range(1,n+1):
        t=t_old+pow(x, i)/math.factorial(i)
        if (t-t_old)<1e-15:
            break
        t_old=t
    return t
```

```python
#creating 1000 nos in between 0 and 100
x=np.linspace(0,100,1000)
#plotting for your_exp
y_mine=[your_exp(i) for i in x]
plt.plot(x, y_mine, label='Calculated by me')
#plotting for np_exp
y_np=[np.exp(i) for i in x]
plt.plot(x, y_np, label='built in numpy exponential function')
plt.legend()
plt.show()

```
